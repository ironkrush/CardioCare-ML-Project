```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('heart_disease.csv')
```

```python
data
```

```python
data.drop(columns=['education'], inplace=True)
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data['prevalentStroke'] = data['prevalentStroke'].replace(['yes','no'],[1,0])
```

```python
data.prevalentStroke.unique()
```

```python
data['Heart_ stroke'] = data['Heart_ stroke'].replace(['yes','No'],[1,0])
```

```python
data
```

```python
data_input = data.drop(columns=['Heart_ stroke'])
data_output = data['Heart_ stroke']
```

```python
data_output.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=2)
```

```python
data_input, data_output = sm.fit_resample(data_input, data_output.ravel())
```

```python
np.unique(data_output, return_counts=True)
#here its showing the count which we have used the imblearn library to resample the data to get the no. of data equal nd done with the help of SMOTE function and in the below line we are checking it worked or not 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_input, data_output, test_size=0.2)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
import pickle as pk
```

```python
pk.dump(model, open('Heart_diease_model.pkl','wb'))
```
